```
# Required imports
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```
# Required files
!wget https://images.pexels.com/photos/1020315/pexels-photo-1020315.jpeg -O 1.jpg
```

```
## Read the image
# cv.imread(path_to_image, flag)
# flag - cv.IMREAD_COLOR - Loads color image, without alpha channel, Default (-1)
# flag - cv.IMREAD_GRAYSCALE - Loads color image as grayscale (0)
# flag - cv.IMREAD_UNCHANGED - Loads color image with alpha channel (1)

img = cv.imread('./1.jpg', 1)
```

```
# Print the shape of the image
img.shape
```

```
# Print the image (Buggy in notebooks env)
cv.imshow('Image', img)
```

```
# Print the image (using plt)
plt.imshow(img)
plt.axis('off')
plt.show()
```

```
# Images are made of 3 channels, RGB, lets seperate them
R = img[:, :, 0]
G = img[:, :, 1]
B = img[:, :, 2]

fig, axes = plt.subplots(1, 3, figsize=(20, 20))
axes[0].imshow(R)
axes[1].imshow(G)
axes[2].imshow(B)
plt.show()
```

```
R.shape
```

```
# Update the R channel and put the channels together
R = np.random.random_sample(R.shape)

img_ = np.zeros_like(img)
img_[:, :, 0] = R
img_[:, :, 1] = G
img_[:, :, 2] = B
```

```
# show the updated image
plt.imshow(img_)
plt.axis('off')
plt.show()
```

```
# Write the image back
cv.imwrite('new_image.jpg', img_)
```
